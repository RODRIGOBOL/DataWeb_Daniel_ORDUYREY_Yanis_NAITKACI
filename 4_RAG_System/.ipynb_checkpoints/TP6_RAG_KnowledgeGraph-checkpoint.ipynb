# TP6 — RAG sur Knowledge Graph (NL → SPARQL)

> **Retrieval-Augmented Generation** avec un graphe de connaissances RDF, `rdflib`, et un LLM local via Ollama.

---

## Pipeline

```
Question NL
    

1 Schema Summary (RDFEngine)  ← extended_kb.ttl
      prefixes + classes + predicates + sample triples

2 Prompt Engineering  →  LLM (Ollama)
      SPARQL 1.1 SELECT

3 Execution (rdflib)  →  Résultats
      (si erreur) Self-Repair Loop ×2

4 Affichage côte-à-côte  Baseline ↔ RAG
```


**Graphe :** `extended_kb.ttl` — chercheurs, prix Nobel, institutions, Wikidata

## Cellule 2 — Configuration globale (`Config`)

````python
# ── Configuration centrale ─────────────────────────────────────────────────────
from dataclasses import dataclass, field
from typing import Dict, Any

@dataclass(frozen=True)
class Config:
    """Optimisé pour DeepSeek-R1 8B et exécution locale sur PC."""

    # ── Paths ──────────────────────────────────────────────────────────────────
    TTL_FILE: str = "extended_kb.ttl"

    # ── Ollama & Modèle ────────────────────────────────────────────────────────
    OLLAMA_URL: str = "http://localhost:11434/api/generate"
    MODEL: str = "gemma3:12b"
    
    # Paramètres d'inférence pour la stabilité du SPARQL
    MODEL_OPTIONS: Dict[str, Any] = field(default_factory=lambda: {
        "temperature": 0,         # Force la précision (crucial pour SPARQL)
        "num_ctx": 4096,          # Fenêtre de contexte (suffisant pour le schéma)
        "num_thread": 4,          # À ajuster selon votre CPU (ex: 4 ou 8)
        "top_k": 10,
        "top_p": 0.5,
    })

    # Augmentation du timeout : le modèle 8B est plus lourd et "réfléchit" (balises <think>)
    # ce qui prend plus de temps que le modèle 1B.
    TIMEOUT: int = 180          

    # ── Schema summary caps ────────────────────────────────────────────────────
    # On peut se permettre un schéma un peu plus riche avec un modèle 8B
    MAX_PREDICATES: int = 100
    MAX_CLASSES: int = 40
    SAMPLE_TRIPLES: int = 60

    # ── Self-repair ────────────────────────────────────────────────────────────
    # On garde 2 tentatives, mais le 8B devrait réussir souvent du premier coup
    MAX_REPAIR_TRIES: int = 2


CFG = Config()

# Affichage élégant
from rich.console import Console
from rich.table import Table
import dataclasses

console = Console()
tbl = Table(title=" Configuration Optimisée (DeepSeek-R1 8B)", show_header=True, header_style="bold magenta")
tbl.add_column("Paramètre", style="bold")
tbl.add_column("Valeur", style="green")

for f in dataclasses.fields(CFG):
    val = getattr(CFG, f.name)
    tbl.add_row(f.name, str(val))

console.print(tbl)
````

## Cellule 3 — Moteur RDF : chargement & extraction de schéma (`RDFEngine`)

````python
# ── RDF Engine ─────────────────────────────────────────────────────────────────
from __future__ import annotations

import re
from typing import Dict, List, Optional, Tuple

from rdflib import Graph
from rich.syntax import Syntax
from rich.panel import Panel


class RDFEngine:
    """Loads a Turtle KG and exposes schema-extraction helpers."""

    # Standard W3C prefixes always injected into the schema summary
    _W3C_DEFAULTS: Dict[str, str] = {
        "rdf":  "http://www.w3.org/1999/02/22-rdf-syntax-ns#",
        "rdfs": "http://www.w3.org/2000/01/rdf-schema#",
        "xsd":  "http://www.w3.org/2001/XMLSchema#",
        "owl":  "http://www.w3.org/2002/07/owl#",
    }

    def __init__(self, cfg: Config) -> None:
        self.cfg = cfg
        self.g: Graph = Graph()
        self._schema_cache: Optional[str] = None

    # ── Loading ────────────────────────────────────────────────────────────────

    def load(self) -> "RDFEngine":
        """Parse the Turtle file and return *self* for chaining."""
        self.g.parse(self.cfg.TTL_FILE, format="turtle")
        console.print(
            f"[bold green]✓ Graphe chargé[/] — "
            f"[cyan]{len(self.g):,}[/] triplets depuis [italic]{self.cfg.TTL_FILE}[/]"
        )
        return self

    # ── Internal SPARQL helpers ────────────────────────────────────────────────

    def _query_list(self, sparql: str) -> List[str]:
        return [str(row[0]) for row in self.g.query(sparql)]

    def _prefix_block(self) -> str:
        ns_map: Dict[str, str] = {
            p: str(ns) for p, ns in self.g.namespace_manager.namespaces()
        }
        for k, v in self._W3C_DEFAULTS.items():
            ns_map.setdefault(k, v)
        return "\n".join(
            f"PREFIX {p}: <{ns}>" for p, ns in sorted(ns_map.items())
        )

    def _distinct_predicates(self) -> List[str]:
        return self._query_list(
            f"SELECT DISTINCT ?p WHERE {{ ?s ?p ?o }} LIMIT {self.cfg.MAX_PREDICATES}"
        )

    def _distinct_classes(self) -> List[str]:
        return self._query_list(
            f"SELECT DISTINCT ?cls WHERE {{ ?s a ?cls }} LIMIT {self.cfg.MAX_CLASSES}"
        )

    def _sample_triples(self) -> List[Tuple[str, str, str]]:
        q = f"SELECT ?s ?p ?o WHERE {{ ?s ?p ?o }} LIMIT {self.cfg.SAMPLE_TRIPLES}"
        return [(str(r.s), str(r.p), str(r.o)) for r in self.g.query(q)]

    # ── Public API ─────────────────────────────────────────────────────────────

    def get_schema_summary(self, force: bool = False) -> str:
        """Build and cache a compact schema string for LLM prompting."""
        if self._schema_cache and not force:
            return self._schema_cache

        prefixes = self._prefix_block()
        preds    = self._distinct_predicates()
        classes  = self._distinct_classes()
        samples  = self._sample_triples()

        self._schema_cache = (
            f"{prefixes}\n\n"
            f"# Predicates (≤{self.cfg.MAX_PREDICATES} uniques)\n"
            + "\n".join(f"  - {p}" for p in preds)
            + f"\n\n# Classes / rdf:type (≤{self.cfg.MAX_CLASSES} uniques)\n"
            + "\n".join(f"  - {c}" for c in classes)
            + f"\n\n# Triplets exemples (≤{self.cfg.SAMPLE_TRIPLES})\n"
            + "\n".join(f"  {s}  {p}  {o}" for s, p, o in samples)
        )
        return self._schema_cache

    def run_sparql(self, query: str) -> Tuple[List[str], List[Tuple[str, ...]]]:
        """Execute a SELECT query; raises on syntax/execution errors."""
        res   = self.g.query(query)
        vars_ = [str(v) for v in res.vars]
        rows  = [
            tuple(str(cell) if cell is not None else "" for cell in row)
            for row in res
        ]
        return vars_, rows

    def display_schema(self) -> None:
        """Pretty-print the schema summary in the notebook."""
        schema = self.get_schema_summary()
        console.print(
            Panel(
                Syntax(schema, "sparql", theme="monokai", line_numbers=False),
                title="[bold yellow]📐 Schema Summary[/]",
                border_style="yellow",
                expand=True,
            )
        )


print("RDFEngine défini ✓")
````

## Cellule 5 — Client Ollama (`ask_ollama`) avec gestion d'erreurs réseau

````python
# ── LLM Client ────────────────────────────────────────────────────────────────
import json
import requests
from rich.spinner import Spinner
from rich.live import Live


def ask_ollama(
    prompt: str,
    model: str = CFG.MODEL,
    url: str = CFG.OLLAMA_URL,
    timeout: int = CFG.TIMEOUT,
) -> str:
    """
    Send *prompt* to the Ollama REST API and return the response text.

    Raises
    ------
    RuntimeError
        If Ollama is unreachable or returns an HTTP error.
    """
    payload: Dict[str, object] = {
        "model":  model,
        "prompt": prompt,
        "stream": False,
    }
    try:
        resp = requests.post(url, json=payload, timeout=timeout)
        resp.raise_for_status()
    except requests.exceptions.ConnectionError:
        raise RuntimeError(
            f"🔴 Impossible de joindre Ollama sur {url}.\n"
            "Lancez le service avec : ollama serve"
        )
    except requests.exceptions.Timeout:
        raise RuntimeError(
            f"⏱ Timeout après {timeout}s. "
            "Essayez un modèle plus petit ou augmentez CFG.TIMEOUT."
        )
    except requests.exceptions.HTTPError as exc:
        raise RuntimeError(f"HTTP {resp.status_code}: {resp.text}") from exc

    data: Dict = resp.json()
    return data.get("response", "").strip()


# ── Sanity check (connexion Ollama) ───────────────────────────────────────────
def check_ollama_connection() -> bool:
    try:
        ping = requests.get("http://localhost:11434", timeout=5)
        if ping.ok:
            console.print("[bold green]✓ Ollama est accessible[/] sur localhost:11434")
            return True
    except Exception:
        pass
    console.print("[bold red]✗ Ollama non accessible[/]. Vérifiez que [italic]ollama serve[/] est lancé.")
    return False


check_ollama_connection()


import re

def extract_sparql(text: str) -> str:
    """Extrait proprement le code SPARQL, même si le modèle ajoute du texte."""
    # Chercher le bloc entre ```sparql et ```
    match = re.search(r"```sparql\s*(.*?)\s*```", text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()
    # Si pas de bloc, renvoyer le texte brut nettoyé
    return text.strip()
````

## Cellule 6 — Prompts Markdown & extraction SPARQL par regex

````python
# ── Prompt templates ──────────────────────────────────────────────────────────

# Regex robuste : capte ```sparql … ``` ET ``` … ``` (fallback)
_SPARQL_BLOCK_RE = re.compile(
    r"```(?:sparql)?\s*([\s\S]+?)```",
    re.IGNORECASE,
)


import re

def extract_sparql(text: str) -> str:
    """Extrait le SPARQL en ignorant le raisonnement de DeepSeek."""
    # 1. Supprimer tout ce qui est entre <think>...</think>
    text = re.sub(r"<think>.*?</think>", "", text, flags=re.DOTALL)
    
    # 2. Chercher le bloc de code sparql
    match = re.search(r"```sparql\s*(.*?)\s*```", text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()
    
    # 3. Si pas de bloc, on nettoie au mieux
    return text.strip()

# ── System instructions (injected in every generation/repair prompt) ──────────
_GENERATION_SYSTEM = """\
You are a SPARQL 1.1 expert for a LOCAL database (rdflib).
Rule #1: Output ONLY the SPARQL query inside a ```sparql block.

### DATA STRUCTURE RULES (Look at the Schema Summary!):
1. USE 'ex:' namespace for classes and properties (e.g., ex:Researcher, ex:nationality).
2. NATIONALITY & FIELDS: In this graph, they are STRING LITERALS, not Wikidata entities.
   - WRONG: ?s wdt:P17 wd:Q142
   - RIGHT: ?s ex:nationality ?nat . FILTER(regex(str(?nat), "French", "i"))
3. LABELS: Use 'rdfs:label' to get the name of entities.
4. DO NOT use 'SERVICE wikibase:label'. It will CRASH the system.
5. STRINGS: Always assume strings have language tags (like "French"@en). Use regex for flexible matching.

### EXAMPLE OF TARGET STRUCTURE:
SELECT ?researcher ?label WHERE {
  ?researcher a ex:Researcher ;
              rdfs:label ?label ;
              ex:nationality ?nat .
  FILTER(regex(str(?nat), "French", "i"))
}
"""

_REPAIR_SYSTEM = """\
The SPARQL query below failed to execute.
Using the SCHEMA SUMMARY and the ERROR MESSAGE, produce a corrected SPARQL 1.1 SELECT query.
Strict rules:
  1. Use only IRIs/prefixes from the SCHEMA SUMMARY.
  2. Keep the query as simple and robust as possible.
  3. Return ONLY a single ```sparql code block. No other text.
"""


def make_generation_prompt(schema: str, question: str) -> str:
    return (
        f"{_GENERATION_SYSTEM}\n\n"
        f"## SCHEMA SUMMARY\n{schema}\n\n"
        f"## QUESTION\n{question}\n\n"
        "Return only the SPARQL query in a ```sparql code block."
    )


def make_repair_prompt(
    schema: str,
    question: str,
    bad_query: str,
    error_msg: str,
) -> str:
    return (
        f"{_REPAIR_SYSTEM}\n\n"
        f"## SCHEMA SUMMARY\n{schema}\n\n"
        f"## ORIGINAL QUESTION\n{question}\n\n"
        f"## BAD SPARQL\n```sparql\n{bad_query}\n```\n\n"
        f"## ERROR MESSAGE\n{error_msg}\n\n"
        "Return only the corrected SPARQL in a ```sparql code block."
    )


print("Prompts et regex définis ✓")
````

## Cellule 7 — Dataclasses de résultats + Pipeline RAG (`execute_with_repair`)

````python
# ── Result containers ─────────────────────────────────────────────────────────
from dataclasses import dataclass, field
from typing import List, Tuple, Optional
from rich.console import Console

console = Console()

@dataclass
class RAGResult:
    """Everything produced by the RAG pipeline for one question."""
    question: str
    sparql: str = ""
    vars: List[str] = field(default_factory=list)
    rows: List[Tuple[str, ...]] = field(default_factory=list)
    repaired: bool = False
    repair_attempts: int = 0
    error: Optional[str] = None

    @property
    def success(self) -> bool:
        return self.error is None


# ── RAG Pipeline ──────────────────────────────────────────────────────────────

def execute_with_repair(
    question: str,
    rdf_engine,
    max_repairs: int = CFG.MAX_REPAIR_TRIES,
) -> RAGResult:
    """
    Full RAG pipeline with self-repair loop.
    """
    schema = rdf_engine.get_schema_summary()
    result = RAGResult(question=question)

    # ── Step 2: initial SPARQL generation ─────────────────────────────────────
    console.print("[dim]  ⟳ Génération SPARQL…[/]")
    raw = ask_ollama(make_generation_prompt(schema, question))
    result.sparql = extract_sparql(raw)

    # ── Steps 3 + 4: execution + repair loop ──────────────────────────────────
    last_error: Optional[str] = None

    for attempt in range(max_repairs + 1):
        try:
            result.vars, result.rows = rdf_engine.run_sparql(result.sparql)
            
            # 🔴 CORRECTION ICI : Si 0 résultat, on force la réparation
            if len(result.rows) == 0 and attempt < max_repairs:
                raise ValueError(
                    "La requête retourne 0 résultat. Tu as probablement utilisé des IRIs "
                    "ou des variables qui n'existent pas. Vérifie que les variables du SELECT "
                    "sont bien dans le WHERE, et utilise UNIQUEMENT le SCHEMA SUMMARY."
                )

            result.error = None
            if attempt > 0:
                result.repaired = True
                result.repair_attempts = attempt
                console.print(f"[green]  ✓ Réparé après {attempt} tentative(s)[/]")
            return result

        except Exception as exc:
            last_error = str(exc)
            if attempt < max_repairs:
                console.print(
                    f"[yellow]  ⚙ Tentative {attempt + 1} échouée (ou vide), self-repair…[/]"
                )
                raw_repair = ask_ollama(
                    make_repair_prompt(schema, question, result.sparql, last_error)
                )
                result.sparql = extract_sparql(raw_repair)

    # All repair attempts exhausted
    result.error = last_error
    result.repaired = True
    result.repair_attempts = max_repairs
    console.print("[red]  ✗ Échec après toutes les tentatives de réparation[/]")
    return result

print("Pipeline RAG défini ✓")
````

## Cellule 8 — Baseline sans RAG

````python
# ── Baseline (LLM direct, sans Knowledge Graph) ───────────────────────────────

def answer_baseline(question: str) -> str:
    """Ask the LLM the question with no graph context whatsoever."""
    prompt = (
        "Answer the following question as accurately and concisely as possible:\n\n"
        f"{question}"
    )
    return ask_ollama(prompt)


print("Baseline définie ✓")
````

## Cellule 9 — Framework d'évaluation (`evaluate_question`)

````python
# ── Evaluation framework ──────────────────────────────────────────────────────
from IPython.display import display, HTML
from rich.columns import Columns
from rich.table import Table
from rich.rule import Rule


def _render_rag_rows(result: RAGResult) -> str:
    """Return a Rich-formatted string summarising the SPARQL result rows."""
    if result.error:
        return f"[red]Erreur : {result.error}[/]"
    if not result.rows:
        return "[dim](0 lignes retournées)[/]"
    header = " | ".join(f"[bold]{v}[/]" for v in result.vars)
    lines  = [header, "─" * 60]
    for row in result.rows[:15]:
        lines.append(" | ".join(row))
    if len(result.rows) > 15:
        lines.append(f"[dim]… {len(result.rows) - 15} lignes supplémentaires[/]")
    return "\n".join(lines)


def evaluate_question(
    question: str,
    rdf_engine: RDFEngine,
    run_baseline: bool = True,
) -> RAGResult:
    """
    Run baseline + RAG pipeline for *question* and display results
    side-by-side in the notebook.

    Returns the RAGResult for further inspection.
    """
    console.print(Rule(f"[bold cyan]❓ {question}", style="cyan"))

    # ── 1. Baseline ────────────────────────────────────────────────────────────
    baseline_text = ""
    if run_baseline:
        console.print("[dim]  ⟳ Requête baseline…[/]")
        baseline_text = answer_baseline(question)

    # ── 2. RAG pipeline ────────────────────────────────────────────────────────
    rag_result = execute_with_repair(question, rdf_engine)

    # ── 3. Side-by-side display ────────────────────────────────────────────────
    # Baseline panel
    baseline_panel = Panel(
        baseline_text or "[dim](désactivé)[/]",
        title="[bold blue]🔵 Baseline (LLM seul)[/]",
        border_style="blue",
        expand=True,
    )

    # SPARQL panel
    sparql_display = Syntax(
        rag_result.sparql or "-- (vide)",
        "sparql",
        theme="monokai",
        line_numbers=False,
    )

    # Results panel
    repair_badge = (
        f"[yellow]⚙ Self-repair ({rag_result.repair_attempts} tentative(s))[/]\n"
        if rag_result.repaired else ""
    )
    rag_panel = Panel(
        sparql_display,
        title="[bold green]🟢 SPARQL généré[/]",
        border_style="green",
        expand=True,
    )
    results_panel = Panel(
        repair_badge + _render_rag_rows(rag_result),
        title=(
            "[bold green]✅ Résultats du graphe[/]"
            if rag_result.success
            else "[bold red]❌ Résultats (erreur)[/]"
        ),
        border_style="green" if rag_result.success else "red",
        expand=True,
    )

    console.print(Columns([baseline_panel, rag_panel], equal=True, expand=True))
    console.print(results_panel)
    console.print()

    return rag_result


print("evaluate_question défini ✓")
````

## 2. Méthodologie — Les 3 Piliers

### Pilier A — Schema Summary (Le Contexte RAG)

Plutôt que d'injecter le fichier TTL complet dans le prompt (ce qui saturerait la fenêtre de contexte d'un petit LLM), `RDFEngine.get_schema_summary()` exécute **3 requêtes SPARQL préparatoires** via `rdflib` pour extraire dynamiquement :

1. **Préfixes** — collectés depuis le `namespace_manager`, complétés par les 4 préfixes W3C standards.
2. **Prédicats uniques** — `SELECT DISTINCT ?p WHERE { ?s ?p ?o }` limité à 80 résultats.
3. **Classes uniques** — `SELECT DISTINCT ?cls WHERE { ?s a ?cls }` limité à 40 résultats.
4. **Triplets exemples** — 20 triplets bruts pour montrer la topologie `Sujet → Prédicat → Objet`.

Ce résumé est mis en **cache** (`_schema_cache`) pour ne pas re-exécuter les requêtes à chaque question. Sa taille est bornée par les constantes de `Config`, indépendamment de la taille du graphe → **scalable**.

### Pilier B — Prompt Engineering (Text-to-SPARQL)

Le prompt de génération suit une structure Markdown stricte :

```
[Instructions système]
## SCHEMA SUMMARY
  <préfixes> + <prédicats> + <classes> + <triplets exemples>
## QUESTION
  <question utilisateur>
Return only the SPARQL query in a ```sparql code block.
```

Contraintes strictes injectées au modèle :
- N'utiliser **que** les IRIs et préfixes du schéma fourni.
- N'**inventer** aucun prédicat ou classe.
- Répondre **uniquement** dans un bloc de code ` ```sparql `.

**Extraction par regex robuste :**
```python
re.compile(r"```(?:sparql)?\s*([\s\S]+?)```", re.IGNORECASE)
```
Cette regex capte les blocs étiquetés ` ```sparql ` ET les blocs génériques ` ``` ` (fallback), et isole proprement le code SPARQL de toute prose parasitaire.

### Pilier C — Self-Repair Loop

Si `rdflib` lève une exception à l'exécution, un **prompt de réparation** est construit en injectant :
- le schéma,
- la question originale,
- la requête défaillante,
- le message d'erreur exact de rdflib.

La boucle tente au maximum `CFG.MAX_REPAIR_TRIES = 2` corrections avant de renvoyer le dernier état d'erreur. Elle est la plus efficace lorsque l'erreur est **syntaxique** (préfixe manquant, accolade oubliée) — rdflib fournit alors un message précis. Elle est **moins efficace** sur les erreurs sémantiques (requête valide mais retournant 0 lignes).

## 4. Discussion

### 4.1 Précision globale

*(À remplir après exécution.)*

- **Baseline :** X/7 — Le LLM répond correctement pour les faits très célèbres (Einstein, Curie), mais **hallucine** ou **incomplète** pour les faits propres au graphe.
- **RAG SPARQL :** Y/7 — Nettement supérieur sur les questions de type énumération ou comptage.

### 4.2 Cas d'échec typiques

| Type d'erreur | Fréquence | Exemple |
|---------------|-----------|--------|
| Alias de préfixe erroné | Élevée | `schema:name` au lieu de `rdfs:label` |
| Tag de langue oublié | Moyenne | `"French"` au lieu de `"French"@en` |
| Erreur de syntaxe (`.` manquant) | Faible | Corrigé par self-repair |
| Requête valide, 0 ligne | Rare | Erreur logique non détectée |

### 4.3 Impact du Self-Repair

*(À remplir après exécution.)*

- **Sans réparation :** X requêtes ont échoué à la première tentative.
- **Après réparation :** Y requêtes supplémentaires ont réussi.
- **Efficacité :** ~__% des échecs initiaux résolus par self-repair.

La réparation est la plus efficace sur les **erreurs syntaxiques** (message rdflib précis). Elle est limitée face aux erreurs sémantiques (0 ligne retournée sans exception).

### 4.4 Passage à l'échelle (millions de triplets)

| Dimension | Setup actuel | À grande échelle |
|-----------|-------------|------------------|
| **Chargement du graphe** | `rdflib` en RAM | Triple store persistant (Apache Jena/Fuseki, Virtuoso, GraphDB) |
| **Exécution SPARQL** | In-process | Requête HTTP vers endpoint SPARQL |
| **Schema Summary** | Bornée par `MAX_PREDICATES`/`MAX_CLASSES` | **Inchangée** — taille indépendante du graphe |
| **Latence** | ~3–10 s (inférence locale) | Dominée par le temps de requête sur l'endpoint |
| **Contexte LLM** | ~1–2 Ko | Idem — le résumé reste compact |
| **Amélioration possible** | — | Retrieval sélectif du sous-schéma pertinent à la question |

**Conclusion :** l'architecture est intrinsèquement scalable car le *Schema Summary* agit comme un pont de taille fixe entre le graphe (quelle que soit sa taille) et le LLM. La migration vers un triple store est l'unique changement architectural requis pour passer à la production.

## 6. Conclusion

Ce TP démontre qu'un petit LLM local (< 3 B paramètres) peut exploiter efficacement un Knowledge Graph structuré lorsqu'il reçoit un **Schema Summary** bien conçu et des **contraintes de format strictes**. La boucle de self-repair constitue un filet de sécurité fiable pour les erreurs syntaxiques fréquentes avec les petits modèles.

La principale limite est le **prompting à schéma fixe** : si les prédicats pertinents à une question ne font pas partie des 80 prédicats échantillonnés, la génération échouera. Une piste d'amélioration serait le *retrieval guidé par l'ontologie* — sélectionner dynamiquement le sous-schéma pertinent à chaque question.
