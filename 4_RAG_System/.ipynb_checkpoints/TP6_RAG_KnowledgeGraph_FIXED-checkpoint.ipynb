# TP6 — RAG on Knowledge Graph (NL → SPARQL)> **Retrieval-Augmented Generation** with an RDF knowledge graph, `rdflib`, and a local LLM via Ollama.---## OverviewThis notebook implements a **Natural Language to SPARQL** pipeline in four stages:```User Question (NL)        │        ▼① Schema Summary (RDFEngine)   ← extended_kb.ttl    prefixes + classes + predicates + sample triples + literal values        │        ▼② Prompt Engineering  ──►  LLM (Ollama / gemma3:12b)    structured prompt with schema context        │        ▼③ SPARQL Execution (rdflib)   ──►  Results    with Self-Repair Loop ×2        │        ▼④ Side-by-side display   Baseline ↔ RAG```**Knowledge Graph:** `extended_kb.ttl` — researchers, Nobel prizes, institutions, Wikidata links**Key design choices:**- The `ex:` namespace (`http://example.org/private#`) is used for all local entities and properties.- Nationality, field of study, and awards are **string literals** with `@en` language tags — not Wikidata entity IRIs.- The schema summary includes **actual literal values** so the LLM never needs to guess data formats.

---## Cell 2 — Global Configuration (`Config`)All tunable parameters are centralised in a single frozen dataclass.| Parameter | Role ||-----------|------|| `TTL_FILE` | Path to the Turtle knowledge graph || `MODEL` | Ollama model name — `gemma3:12b` is the best balance of speed and SPARQL accuracy || `TIMEOUT` | HTTP timeout for Ollama calls (seconds) || `SAMPLE_TRIPLES` | Number of example triples injected in the schema summary (only `ex:` subjects) || `MAX_REPAIR_TRIES` | How many self-repair attempts before giving up |> **Why `temperature: 0`?** SPARQL is a formal language — we want the most deterministic output possible, not creative variations.

````python
from dataclasses import dataclass, fieldfrom typing import Dict, Any@dataclass(frozen=True)class Config:    """Central configuration — optimised for gemma3:12b on a local machine."""    # ── Paths ──────────────────────────────────────────────────────────────────    TTL_FILE: str = "extended_kb.ttl"    # ── Ollama & Model ─────────────────────────────────────────────────────────    OLLAMA_URL: str = "http://localhost:11434/api/generate"    MODEL: str = "gemma3:12b"    MODEL_OPTIONS: Dict[str, Any] = field(default_factory=lambda: {        "temperature": 0,      # Deterministic output — critical for SPARQL correctness        "num_ctx": 4096,       # Context window — enough for the schema summary        "num_thread": 4,       # Adjust to your CPU core count        "top_k": 10,        "top_p": 0.5,    })    TIMEOUT: int = 180         # Seconds — increase if using a larger model    # ── Schema summary caps ────────────────────────────────────────────────────    MAX_PREDICATES: int = 100    MAX_CLASSES: int = 40    SAMPLE_TRIPLES: int = 60   # Only ex: subjects (see RDFEngine below)    # ── Self-repair ────────────────────────────────────────────────────────────    MAX_REPAIR_TRIES: int = 2CFG = Config()# Pretty-print the configurationfrom rich.console import Consolefrom rich.table import Tableimport dataclassesconsole = Console()tbl = Table(title="Active Configuration", show_header=True, header_style="bold magenta")tbl.add_column("Parameter", style="bold")tbl.add_column("Value", style="green")for f in dataclasses.fields(CFG):    tbl.add_row(f.name, str(getattr(CFG, f.name)))console.print(tbl)
````

---## Cell 3 — RDF Engine: Loading & Schema ExtractionThe `RDFEngine` class is the **RAG retriever** of this pipeline. It:1. Loads the Turtle file into an in-memory `rdflib.Graph`2. Extracts a compact **schema summary** to feed the LLM as context### Why a schema summary instead of the full file?The full TTL file is ~52,000 triples — far too large for a small LLM's context window.Instead, `get_schema_summary()` runs 4 lightweight SPARQL queries to extract:- **Prefixes** — so the LLM uses the right namespace shortcuts- **Distinct predicates** — the full vocabulary of the graph- **Distinct classes** — `rdf:type` values- **Example triples** — filtered to `ex:` subjects only (critical fix: avoids showing Wikidata triples first)- **Literal values** — the actual string values stored in key properties (`ex:nationality`, `ex:fieldOfStudy`, etc.)The last point is the most important for correctness: by explicitly listing `'French'@en`, `'British'@en`, etc., the LLM never needs to guess that nationalities are stored as string literals rather than Wikidata entity IRIs like `wd:Q142`.

````python
from __future__ import annotationsimport refrom typing import Dict, List, Optional, Tuplefrom rdflib import Graphfrom rich.syntax import Syntaxfrom rich.panel import Panel# The local namespace — all researchers and properties use this prefixEX_NS = "http://example.org/private#"class RDFEngine:    """Loads a Turtle knowledge graph and provides schema-extraction for RAG prompting."""    # Standard W3C prefixes always injected into the schema summary    _W3C_DEFAULTS: Dict[str, str] = {        "rdf":  "http://www.w3.org/1999/02/22-rdf-syntax-ns#",        "rdfs": "http://www.w3.org/2000/01/rdf-schema#",        "xsd":  "http://www.w3.org/2001/XMLSchema#",        "owl":  "http://www.w3.org/2002/07/owl#",    }    def __init__(self, cfg: Config) -> None:        self.cfg = cfg        self.g: Graph = Graph()        self._schema_cache: Optional[str] = None    # ── Loading ────────────────────────────────────────────────────────────────    def load(self) -> "RDFEngine":        """Parse the Turtle file and return self for chaining."""        self.g.parse(self.cfg.TTL_FILE, format="turtle")        console.print(            f"[bold green]✓ Graph loaded[/] — "            f"[cyan]{len(self.g):,}[/] triples from [italic]{self.cfg.TTL_FILE}[/]"        )        return self    # ── Internal SPARQL helpers ────────────────────────────────────────────────    def _query_list(self, sparql: str) -> List[str]:        return [str(row[0]) for row in self.g.query(sparql)]    def _prefix_block(self) -> str:        ns_map: Dict[str, str] = {            p: str(ns) for p, ns in self.g.namespace_manager.namespaces()        }        for k, v in self._W3C_DEFAULTS.items():            ns_map.setdefault(k, v)        return "\n".join(            f"PREFIX {p}: <{ns}>" for p, ns in sorted(ns_map.items())        )    def _distinct_predicates(self) -> List[str]:        return self._query_list(            f"SELECT DISTINCT ?p WHERE {{ ?s ?p ?o }} LIMIT {self.cfg.MAX_PREDICATES}"        )    def _distinct_classes(self) -> List[str]:        return self._query_list(            f"SELECT DISTINCT ?cls WHERE {{ ?s a ?cls }} LIMIT {self.cfg.MAX_CLASSES}"        )    def _sample_triples(self) -> List[Tuple[str, str, str]]:        """        Returns only triples whose subject is in the ex: namespace.        Why this filter matters:        Without it, the LIMIT 60 query returns Wikidata triples first        (e.g. wd:Q62897 wdt:P166 wd:...), causing the LLM to generate        Wikidata-style SPARQL instead of local ex: queries.        """        q = (            f"SELECT ?s ?p ?o WHERE {{"            f"  ?s ?p ?o ."            f"  FILTER(STRSTARTS(STR(?s), \"{EX_NS}\"))"            f"}} LIMIT {self.cfg.SAMPLE_TRIPLES}"        )        return [(str(r.s), str(r.p), str(r.o)) for r in self.g.query(q)]    def _literal_values_block(self) -> str:        """        Extracts the actual string values stored in key properties.        This is the most important RAG context: by seeing that ex:nationality        contains 'French'@en (not wd:Q142), the LLM generates correct FILTER clauses.        """        props = {            "ex:nationality":    f"SELECT DISTINCT ?v WHERE {{ ?s <{EX_NS}nationality> ?v }} LIMIT 20",            "ex:fieldOfStudy":   f"SELECT DISTINCT ?v WHERE {{ ?s <{EX_NS}fieldOfStudy> ?v }} LIMIT 20",            "ex:wonAward":       f"SELECT DISTINCT ?v WHERE {{ ?s <{EX_NS}wonAward> ?v }} LIMIT 20",            "ex:affiliatedWith": f"SELECT DISTINCT ?v WHERE {{ ?s <{EX_NS}affiliatedWith> ?v }} LIMIT 20",        }        lines = ["# ── ACTUAL LITERAL VALUES (use these in FILTER clauses) ──"]        for prop_label, sparql in props.items():            vals = [str(r[0]) for r in self.g.query(sparql)]            lines.append(f"# {prop_label}: {', '.join(repr(str(v)) for v in vals[:10])}")        return "\n".join(lines)    # ── Public API ─────────────────────────────────────────────────────────────    def get_schema_summary(self, force: bool = False) -> str:        """Build and cache a compact schema string for LLM prompting."""        if self._schema_cache and not force:            return self._schema_cache        prefixes = self._prefix_block()        preds    = self._distinct_predicates()        classes  = self._distinct_classes()        samples  = self._sample_triples()        literals = self._literal_values_block()        self._schema_cache = (            f"{prefixes}\n\n"            f"# Predicates (≤{self.cfg.MAX_PREDICATES} unique)\n"            + "\n".join(f"  - {p}" for p in preds)            + f"\n\n# Classes / rdf:type (≤{self.cfg.MAX_CLASSES} unique)\n"            + "\n".join(f"  - {c}" for c in classes)            + f"\n\n{literals}\n\n"            + f"# Example triples (≤{self.cfg.SAMPLE_TRIPLES}, ex: subjects only)\n"            + "\n".join(f"  {s}  {p}  {o}" for s, p, o in samples)        )        return self._schema_cache    def run_sparql(self, query: str) -> Tuple[List[str], List[Tuple[str, ...]]]:        """Execute a SPARQL SELECT query. Raises on syntax or execution errors."""        res   = self.g.query(query)        vars_ = [str(v) for v in res.vars]        rows  = [            tuple(str(cell) if cell is not None else "" for cell in row)            for row in res        ]        return vars_, rows    def display_schema(self) -> None:        """Pretty-print the schema summary in the notebook."""        schema = self.get_schema_summary()        console.print(            Panel(                Syntax(schema, "sparql", theme="monokai", line_numbers=False),                title="[bold yellow]📐 Schema Summary[/]",                border_style="yellow",                expand=True,            )        )print("RDFEngine defined ✓")
````

---## Cell 5 — Ollama HTTP ClientA minimal wrapper around the Ollama REST API. Key points:- **Streaming is disabled** (`stream: False`) — we wait for the full response before parsing SPARQL.- **Timeout handling** gives a clear error message suggesting a lighter model if Ollama is too slow.- All generation parameters come from `CFG.MODEL_OPTIONS`, keeping this function stateless.

````python
import requestsdef ask_ollama(prompt: str) -> str:    """Send a prompt to the local Ollama server and return the response text."""    payload = {        "model":   CFG.MODEL,        "prompt":  prompt,        "stream":  False,          # Wait for the complete response        "options": CFG.MODEL_OPTIONS,    }    try:        resp = requests.post(CFG.OLLAMA_URL, json=payload, timeout=CFG.TIMEOUT)        resp.raise_for_status()        return resp.json().get("response", "")    except requests.exceptions.Timeout:        raise TimeoutError(            f"Ollama did not respond within {CFG.TIMEOUT}s. "            "Try a lighter model (e.g. gemma3:4b) or increase CFG.TIMEOUT."        )    except requests.exceptions.ConnectionError:        raise ConnectionError(            "Cannot reach Ollama. Make sure it is running: `ollama serve`"        )print("Ollama client defined ✓")
````

---## Cell 6 — Prompt Engineering & SPARQL ExtractionThis cell contains the three most critical components of the pipeline.### 6a — SPARQL Extraction (`extract_sparql`)LLMs are inconsistent about code block formatting. The extractor handles 4 cases:| Case | Example LLM output | Strategy ||------|-------------------|----------|| Labelled code block | ` ```sparql SELECT... ``` ` | Regex group 1 || Generic code block | ` ``` SELECT... ``` ` | Regex group 1 (fallback label) || No backticks | `SELECT ?label WHERE {...}` | Regex anchored on `SELECT` || Garbage / thinking tags | `<think>...</think> SELECT...` | Strip think tags first |### 6b — Generation System PromptThe prompt enforces a strict **local-only** contract with the LLM. The key rule:> *"Never use `SERVICE`, `wikibase:`, `wd:Q*`, `wdt:P*` — these crash rdflib."*This is reinforced with 4 concrete example queries covering the main question types.### 6c — Repair System PromptUsed when `rdflib` raises an exception. Injects the broken query + error message so the LLM can diagnose and fix the issue.

````python
import re# ── SPARQL extraction — robust to inconsistent LLM formatting ─────────────────def extract_sparql(text: str) -> str:    """    Extract the SPARQL query from an LLM response.    Handles 4 cases in order of preference:      1. ```sparql ... ``` labelled code block      2. ``` ... ``` generic code block      3. Raw SELECT ... } without backticks      4. Fallback: return cleaned text    """    # Remove DeepSeek-R1 reasoning tags (no-op for other models)    text = re.sub(r"<think>.*?</think>", "", text, flags=re.DOTALL)    # Case 1 & 2: fenced code block    match = re.search(r"```(?:sparql)?\s*([\s\S]+?)```", text, re.IGNORECASE)    if match:        return match.group(1).strip()    # Case 3: LLM responded directly without fences    direct = re.search(        r"((?:PREFIX\s+\w+[^\n]*\n)*\s*SELECT\s[\s\S]+?\})\s*$",        text.strip(),        re.IGNORECASE,    )    if direct:        return direct.group(1).strip()    # Fallback    return text.strip()# ── Generation system prompt ──────────────────────────────────────────────────# This is the most impactful part of the pipeline.# The LLM is trained heavily on Wikidata SPARQL; we must override that default._GENERATION_SYSTEM = """\You are a SPARQL 1.1 expert for a LOCAL RDF database queried with rdflib (NOT Wikidata).=== ABSOLUTE RULES — VIOLATION = WRONG ANSWER ===1. Output ONLY a ```sparql code block. No explanation, no prose.2. NEVER use: SERVICE, wikibase:, bd:, wd:Q*, wdt:P* — they CRASH rdflib.3. USE ONLY the prefixes and predicates listed in the SCHEMA SUMMARY below.4. The graph uses ex: (http://example.org/private#) for all researchers and properties.5. Nationality, fieldOfStudy, and wonAward are STRING LITERALS with @en language tags.   - CORRECT:  ?s ex:nationality ?nat . FILTER(regex(str(?nat), "French", "i"))   - CORRECT:  ?s ex:fieldOfStudy ?f  . FILTER(regex(str(?f), "Quantum Physics", "i"))   - WRONG:    ?s wdt:P27 wd:Q142         ← DO NOT DO THIS   - WRONG:    ?s ex:nationality "French" ← missing @en tag, use regex instead6. To get a researcher's name: ?s rdfs:label ?label .7. ex:wonAward links to IRIs like ex:Nobel_Prize_in_Chemistry.   To find Chemistry Nobel winners:   ?s ex:wonAward ex:Nobel_Prize_in_Chemistry .   OR: ?s ex:wonAward ?award . FILTER(regex(str(?award), "Chemistry", "i"))8. To find researchers with a Wikidata link:   ?s owl:sameAs ?wd . FILTER(STRSTARTS(STR(?wd), "http://www.wikidata.org"))=== EXAMPLE QUERIES ===# French researchers:SELECT ?label WHERE {  ?s a ex:Researcher ; rdfs:label ?label ; ex:nationality ?nat .  FILTER(regex(str(?nat), "French", "i"))}# Nobel Chemistry winners:SELECT ?label WHERE {  ?s a ex:Researcher ; rdfs:label ?label ;     ex:wonAward ex:Nobel_Prize_in_Chemistry .}# Researchers at CERN:SELECT ?label WHERE {  ?s a ex:Researcher ; rdfs:label ?label ;     ex:affiliatedWith ex:CERN .}# Count distinct fields of study:SELECT (COUNT(DISTINCT ?f) AS ?count) WHERE {  ?s ex:fieldOfStudy ?f .}"""# ── Repair system prompt ──────────────────────────────────────────────────────_REPAIR_SYSTEM = """\The SPARQL query below failed on a LOCAL rdflib graph (NOT Wikidata).Fix it using ONLY the SCHEMA SUMMARY provided.STRICT RULES:  1. NEVER use SERVICE, wikibase:, wd:, wdt: — these CRASH rdflib.  2. Use only ex: namespace for researchers and properties.  3. Use FILTER(regex(...)) for string properties (nationality, fieldOfStudy).  4. Return ONLY a single ```sparql code block."""def make_generation_prompt(schema: str, question: str) -> str:    """Build the full generation prompt: system instructions + schema + question."""    return (        f"{_GENERATION_SYSTEM}\n\n"        f"## SCHEMA SUMMARY\n{schema}\n\n"        f"## QUESTION\n{question}\n\n"        "Return ONLY the SPARQL query in a ```sparql code block."    )def make_repair_prompt(schema: str, question: str, bad_query: str, error_msg: str) -> str:    """Build the repair prompt: system instructions + schema + broken query + error."""    return (        f"{_REPAIR_SYSTEM}\n\n"        f"## SCHEMA SUMMARY\n{schema}\n\n"        f"## ORIGINAL QUESTION\n{question}\n\n"        f"## BAD SPARQL\n```sparql\n{bad_query}\n```\n\n"        f"## ERROR MESSAGE\n{error_msg}\n\n"        "Return only the corrected SPARQL in a ```sparql code block."    )print("Prompts and regex defined ✓")
````

---## Cell 7 — RAG Pipeline with Self-Repair (`execute_with_repair`)The full pipeline in one function:```① Generate SPARQL  →  ask_ollama(generation_prompt)② Execute          →  rdflib.Graph.query(sparql)       │       ├─ Success + rows  →  return result       ├─ Success + 0 rows  →  treat as error, trigger repair       └─ Exception        →  trigger repair③ Repair loop (max CFG.MAX_REPAIR_TRIES times)       │       ├─ Generate fix  →  ask_ollama(repair_prompt)       ├─ Execute again       └─ (repeat until success or exhausted)```**Why trigger repair on 0 rows?**  `rdflib` considers a well-formed query that returns no results a success — but for our evaluation questions, 0 rows always means the LLM hallucinated a predicate or used a wrong entity IRI. Treating it as an error forces a second attempt with explicit feedback.

````python
from dataclasses import dataclass, field as dc_fieldfrom typing import List, Tuple, Optionalfrom rich.console import Consoleconsole = Console()@dataclassclass RAGResult:    """Container for everything produced by the RAG pipeline for one question."""    question: str    sparql:   str = ""    vars:     List[str]           = dc_field(default_factory=list)    rows:     List[Tuple[str, ...]] = dc_field(default_factory=list)    repaired: bool = False    repair_attempts: int = 0    error:    Optional[str] = None    @property    def success(self) -> bool:        return self.error is Nonedef execute_with_repair(    question: str,    rdf_engine: RDFEngine,    max_repairs: int = CFG.MAX_REPAIR_TRIES,) -> RAGResult:    """    Run the full RAG pipeline: generate SPARQL → execute → self-repair if needed.    Returns a RAGResult with the final query, result rows, and repair metadata.    """    schema = rdf_engine.get_schema_summary()    result = RAGResult(question=question)    # Step ①: Initial SPARQL generation    console.print("[dim]  ⟳ Generating SPARQL…[/]")    raw = ask_ollama(make_generation_prompt(schema, question))    result.sparql = extract_sparql(raw)    last_error: Optional[str] = None    for attempt in range(max_repairs + 1):        try:            result.vars, result.rows = rdf_engine.run_sparql(result.sparql)            # Step ②b: 0 rows — treat as semantic error and repair            if len(result.rows) == 0 and attempt < max_repairs:                raise ValueError(                    "Query executed successfully but returned 0 rows. "                    "Likely cause: wrong predicates, wrong entity IRIs, or missing FILTER. "                    "Re-read the SCHEMA SUMMARY literal values and regenerate."                )            # Step ②a: Genuine success            result.error = None            if attempt > 0:                result.repaired = True                result.repair_attempts = attempt                console.print(f"[green]  ✓ Repaired after {attempt} attempt(s)[/]")            return result        except Exception as exc:            last_error = str(exc)            if attempt < max_repairs:                console.print(                    f"[yellow]  ⚙ Attempt {attempt + 1} failed — self-repairing…[/]"                )                # Step ③: Repair                raw_repair = ask_ollama(                    make_repair_prompt(schema, question, result.sparql, last_error)                )                result.sparql = extract_sparql(raw_repair)    # All repair attempts exhausted    result.error = last_error    result.repaired = True    result.repair_attempts = max_repairs    console.print("[red]  ✗ Failed after all repair attempts[/]")    return resultprint("RAG pipeline defined ✓")
````

---## Cell 8 — Baseline (LLM Without Context)The **baseline** asks the LLM the same question with **no knowledge graph context** — just raw parametric memory.This is the control condition for our evaluation. We expect:- ✓ Correct for famous facts (Einstein won Physics Nobel)- ✗ Incomplete or hallucinated for graph-specific facts (which researchers are at CERN?)- ✗ Unable to count or enumerate exhaustively

````python
def answer_baseline(question: str) -> str:    """Ask the LLM the question with zero graph context (pure parametric memory)."""    prompt = (        "Answer the following question as accurately and concisely as possible:\n\n"        f"{question}"    )    return ask_ollama(prompt)print("Baseline defined ✓")
````

---## Cell 9 — Evaluation Framework (`evaluate_question`)`evaluate_question` runs both the baseline and the RAG pipeline for a given question and displays results **side-by-side** in the notebook:```┌─────────────────────────┐  ┌─────────────────────────┐│  🔵 Baseline (LLM only) │  │  🟢 Generated SPARQL     ││  (free text answer)     │  │  (structured query)      │└─────────────────────────┘  └─────────────────────────┘┌─────────────────────────────────────────────────────────┐│  ✅ Results  /  ❌ Error                                  ││  col1 | col2 | col3                                      ││  ...                                                     │└─────────────────────────────────────────────────────────┘```

````python
from rich.table import Tablefrom rich.rule  import Rulefrom rich.columns import Columnsdef _render_rag_rows(result: RAGResult) -> str:    """Format SPARQL result rows as a Rich-formatted string."""    if result.error:        return f"[red]Error: {result.error}[/]"    if not result.rows:        return "[dim](0 rows returned)[/]"    header = " | ".join(f"[bold]{v}[/]" for v in result.vars)    lines  = [header, "─" * 60]    for row in result.rows[:15]:        lines.append(" | ".join(row))    if len(result.rows) > 15:        lines.append(f"[dim]… {len(result.rows) - 15} additional rows[/]")    return "\n".join(lines)def evaluate_question(    question: str,    rdf_engine: RDFEngine,    run_baseline: bool = True,) -> RAGResult:    """    Run baseline + RAG pipeline for *question* and display results side-by-side.    Returns the RAGResult for further inspection or aggregation.    """    console.print(Rule(f"[bold cyan]❓ {question}", style="cyan"))    # ── Baseline ───────────────────────────────────────────────────────────────    baseline_text = ""    if run_baseline:        console.print("[dim]  ⟳ Baseline query…[/]")        baseline_text = answer_baseline(question)    # ── RAG Pipeline ───────────────────────────────────────────────────────────    rag_result = execute_with_repair(question, rdf_engine)    # ── Side-by-side display ───────────────────────────────────────────────────    left = Panel(        baseline_text or "[dim](baseline disabled)[/]",        title="[bold blue]🔵 Baseline (LLM only)[/]",        border_style="blue",    )    right = Panel(        Syntax(rag_result.sparql, "sparql", theme="monokai"),        title="[bold green]🟢 Generated SPARQL[/]",        border_style="green",    )    console.print(Columns([left, right]))    status_color = "green" if rag_result.success and rag_result.rows else "red"    repair_note  = (        f"[yellow]⚙ Self-repair ({rag_result.repair_attempts} attempt(s))[/]\n"        if rag_result.repaired else ""    )    console.print(        Panel(            repair_note + _render_rag_rows(rag_result),            title=f"[bold {status_color}]"                  f"{'✅ Results' if rag_result.success else '❌ Results (error)'}[/]",            border_style=status_color,            expand=True,        )    )    return rag_resultprint("Evaluation framework defined ✓")
````

Methodology — The 3 Pillars### Pillar A — Schema Summary (The RAG Context)Rather than injecting the full TTL file into the prompt (which would overflow a small LLM's context window), `RDFEngine.get_schema_summary()` runs **5 lightweight SPARQL queries** to dynamically extract:1. **Prefixes** — collected from the `namespace_manager`, extended with 4 W3C standards.2. **Unique predicates** — `SELECT DISTINCT ?p WHERE { ?s ?p ?o }` capped at 100.3. **Unique classes** — `SELECT DISTINCT ?cls WHERE { ?s a ?cls }` capped at 40.4. **Example triples** — 60 triples filtered to `ex:` subjects only (prevents Wikidata confusion).5. **Literal values** — actual string values of `ex:nationality`, `ex:fieldOfStudy`, `ex:wonAward`, `ex:affiliatedWith`.The summary is **cached** (`_schema_cache`) to avoid re-running queries on every question. Its size is bounded by the `Config` constants, independent of graph size → **scalable**.### Pillar B — Prompt Engineering (Text-to-SPARQL)The generation prompt follows a strict structure:```[System instructions — 8 absolute rules + 4 example queries]## SCHEMA SUMMARY  <prefixes> + <predicates> + <classes> + <literal values> + <example triples>## QUESTION  <user question>Return ONLY the SPARQL query in a ```sparql code block.```The key constraint enforced in the prompt:> *"NEVER use SERVICE, wikibase:, wd:Q\*, wdt:P\* — they CRASH rdflib."***Robust extraction regex:**```pythonre.search(r"```(?:sparql)?\s*([\s\S]+?)```", text, re.IGNORECASE)```This captures both ` ```sparql ` and generic ` ``` ` blocks, with a fallback for raw `SELECT` responses.### Pillar C — Self-Repair LoopIf `rdflib` raises an exception **or** the query returns 0 rows, a **repair prompt** is constructed with:- the schema summary- the original question- the broken query- the exact error messageThe loop attempts at most `CFG.MAX_REPAIR_TRIES = 2` corrections. It is most effective on **syntactic errors** (missing prefix, wrong bracket) where rdflib provides a precise message. It is less effective on **semantic errors** (valid query, logically wrong results).---## 3. ConclusionThis TP demonstrates that a small local LLM (< 13B parameters) can effectively query a structured Knowledge Graph when given a well-designed **Schema Summary** and **strict format constraints**. The self-repair loop provides a reliable safety net for the frequent syntactic errors made by small models.The main limitation is **fixed-schema prompting**: if the predicates relevant to a question are not among the sampled predicates, generation will fail. A promising improvement would be *ontology-guided retrieval* — dynamically selecting the sub-schema most relevant to each question before prompting.
